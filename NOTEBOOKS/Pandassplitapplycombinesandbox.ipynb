```python
# %load imports.py
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
sns.set()

```

```python
df = pd.read_csv('../DATA/sales_records.csv')
df
```

```python
df.info()
```

```python
df['Profit'] = df['Units Sold'] * (df['Unit Price'] - df['Unit Cost'])
df.head()
```

```python
pd.options.display.float_format = '{:.2f}'.format
regions = df.groupby('Region')
profit_by_region = regions.sum().sort_values(by="Profit", ascending=False)
profit_by_region
```

```python
profit_by_region.style.highlight_max(['Profit','Units Sold', 'Unit Price', 'Unit Cost'])
```
