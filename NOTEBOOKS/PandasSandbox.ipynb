```python
df = pd.read_csv("../DATA/sales_records.csv", parse_dates=['Order Date', 'Ship Date'])
df
```

```python
df["Order Year"] = df['Order Date'].dt.year
df
```

```python
pt = df[["Item Type", "Country", "Unit Price"]].pivot_table(index="Item Type", columns="Country", aggfunc="min")
pt
```

```python
gb = pt = df[["Item Type", "Country", "Units Sold"]].groupby('Item Type').agg("min")
gb
```

```python
pt = df[["Item Type", "Order Year", "Region", "Units Sold"]].pivot_table(index=['Order Year'], columns="Region",  aggfunc="mean")
pt
```

```python
pt = df[["Order Year", "Units Sold"]].pivot_table(index="Order Year", aggfunc=["min", "max", "mean", "median"])
pt
```

```python
chi = pd.read_csv('../DATA/city-of-chicago-salaries.csv')
chi.columns = 'name position department salary'.split()
chi
```

Convert salary column from string to float after removing leading '$'.

```python
chi['salary'] = chi['salary'].str[1:].astype(float)
chi
```

Calculate average salary per department

```python
pt = chi[["department", "position", "salary"]].pivot_table(index=["department", "position"]).sort_values(by=["department", "salary"])
pt
```
