# Pandas Indexing Overview

```python
# %load imports.py
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
sns.set()

```

```python
values = [  # sample data
    [100, 110, 120, 130, 140],
    [200, 210, 220, 230, 240],
    [300, 310, 320, 330, 340],
    [400, 410, 420, 430, 440],
    [500, 510, 520, 530, 540],
    [600, 610, 620, 630, 640],
]
df = pd.DataFrame(
    values,
    index = ['a', 'b', 'c', 'd', 'e', 'f'],
    columns = ['alpha', 'beta', 'gamma', 'delta', 'epsilon'],
)
df
```
